# Task 1: Load and Explore the Dataset

```python
import pandas as pd
from sklearn.datasets import load_iris

# -------------------------
# Step 1: Load iris dataset from sklearn
# -------------------------
iris = load_iris()
```

```python
print(iris)
```

```python
# Step 2: Convert to Pandas DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["species"] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
df.head()
```

```python
# -------------------------
# Step 3: Inspect data
# -------------------------
print("First 5 rows:")
print(df.head(), "\n")

print("Dataset Info:")
print(df.info(), "\n")

print("Missing Values:")
print(df.isnull().sum(), "\n")

# -------------------------
# Step 4 (optional): Save to CSV
# -------------------------
df.to_csv("iris.csv", index=False)
print("✅ Iris dataset saved as iris.csv")
```

```python
# -------------------------
# Step 1: Load dataset
# -------------------------

df = pd.read_csv("iris.csv")
```

```python
# -------------------------
# Step 2: Display first few rows
# -------------------------
print("First 5 rows of the dataset:")
print(df.head(), "\n")
```

```python
# -------------------------
# Step 3: Inspect structure
# -------------------------
print("Dataset Info:")
print(df.info(), "\n")

print("Summary Statistics:")
print(df.describe(), "\n")
```

```python
# -------------------------
# Step 4: Check for missing values
# -------------------------
print("Missing Values in Each Column:")
print(df.isnull().sum(), "\n")
```

```python


# -------------------------
# Step 5: Clean dataset
# -------------------------
# Option B: Fill missing values (example: fill with mean for numeric cols)
# df_cleaned = df.fillna(df.mean(numeric_only=True))

# Option A: Drop missing values
df_cleaned = df.dropna()


print("After Cleaning:")
print(df_cleaned.info())
```

# Task 2: Basic Data Analysis

```python
# -------------------------------
# 1. Basic statistics
# -------------------------------
print("\nBasic Statistics (describe):")
print(df_cleaned.describe())
```

```python
# -------------------------------
# 2. Grouping by species
# -------------------------------
print("\nMean values by species:")
grouped = df.groupby("species").mean()
```

```python
print(grouped)
```

# Task 3: Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ----------------------------
# 1. Line Chart (simulated time-series using index)
# ----------------------------
plt.figure(figsize=(8, 5))
plt.plot(df_cleaned.index, df_cleaned['sepal length (cm)'], label="Sepal Length", color='blue')
plt.plot(df_cleaned.index, df_cleaned['petal length (cm)'], label="Petal Length", color='red')
plt.title("Line Chart: Sepal & Petal Length over Samples")
plt.xlabel("Sample Index (simulated time)")
plt.ylabel("Length (cm)")
plt.legend()
plt.show()
```

```python
# ----------------------------
# 2. Bar Chart (average petal length per species)
# ----------------------------
plt.figure(figsize=(6, 4))
sns.barplot(x="species", y="petal length (cm)", data=df_cleaned, ci=None, palette="Set2")
plt.title("Average Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.show()
```

```python
# ----------------------------
# 3. Histogram (distribution of sepal length)
# ----------------------------
plt.figure(figsize=(6, 4))
plt.hist(df_cleaned['sepal length (cm)'], bins=15, color='skyblue', edgecolor='black')
plt.title("Histogram: Sepal Length Distribution")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
# ----------------------------
# 4. Scatter Plot (sepal length vs petal length)
# ----------------------------
plt.figure(figsize=(6, 4))
sns.scatterplot(x="sepal length (cm)", y="petal length (cm)",
                hue="species", style="species", palette="deep", data=df_cleaned)
plt.title("Scatter Plot: Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend(title="Species")
plt.show()
```

# SUB_PLOTTING

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# ----------------------------
# 1. Line Chart
# ----------------------------
axes[0, 0].plot(df_cleaned.index, df_cleaned['sepal length (cm)'], label="Sepal Length", color='blue')
axes[0, 0].plot(df_cleaned.index, df_cleaned['petal length (cm)'], label="Petal Length", color='red')
axes[0, 0].set_title("Line Chart: Sepal & Petal Length over Samples")
axes[0, 0].set_xlabel("Sample Index (simulated time)")
axes[0, 0].set_ylabel("Length (cm)")
axes[0, 0].legend()

# ----------------------------
# 2. Bar Chart
# ----------------------------
sns.barplot(x="species", y="petal length (cm)", data=df_cleaned, ci=None, palette="Set2", ax=axes[0, 1])
axes[0, 1].set_title("Average Petal Length by Species")
axes[0, 1].set_xlabel("Species")
axes[0, 1].set_ylabel("Petal Length (cm)")

# ----------------------------
# 3. Histogram
# ----------------------------
axes[1, 0].hist(df_cleaned['sepal length (cm)'], bins=15, color='skyblue', edgecolor='black')
axes[1, 0].set_title("Histogram: Sepal Length Distribution")
axes[1, 0].set_xlabel("Sepal Length (cm)")
axes[1, 0].set_ylabel("Frequency")

# ----------------------------
# 4. Scatter Plot
# ----------------------------
sns.scatterplot(x="sepal length (cm)", y="petal length (cm)",
                hue="species", style="species", palette="deep",
                data=df_cleaned, ax=axes[1, 1])
axes[1, 1].set_title("Scatter Plot: Sepal vs Petal Length")
axes[1, 1].set_xlabel("Sepal Length (cm)")
axes[1, 1].set_ylabel("Petal Length (cm)")

# ----------------------------
# Adjust layout and show
# ----------------------------
plt.tight_layout()
plt.show()
```
